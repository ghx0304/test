```python
import numpy as np
from matplotlib import  pyplot as plt

us_file_path = "./youtube_video_data/US_video_data_numbers.csv"
uk_file_path = "./youtube_video_data/GB_video_data_numbers.csv"

# t1 = np.loadtxt(us_file_path,delimiter=",",dtype="int",unpack=True)
t_us = np.loadtxt(us_file_path,delimiter=",",dtype="int")
print(t_us.shape)

# 取美国的电影的评论的数据
t_us_comments = t_us[:,-1]

#选择比5000小的数据
t_us_comments = t_us_comments[t_us_comments<=5000]

print(t_us_comments.shape)

#再看最小值和最大值
print(t_us_comments.max(),t_us_comments.min())
```

```python
# 输出数据集的形状
print(t_us_comments.shape)

d = 100

# 计算bin的数量
bin_nums = (t_us_comments.max() - t_us_comments.min()) // d

# 绘图
plt.figure(figsize=(20, 8), dpi=80)

# 绘制直方图
plt.hist(t_us_comments, bin_nums)

# 添加x轴和y轴的0线
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# 设置x轴和y轴的范围，确保0点对齐
plt.xlim(left=min(t_us_comments.min(), 0), right=t_us_comments.max())
plt.ylim(bottom=min(0, 0), top=300)

# 设置x轴显示更多刻度
x_ticks = np.arange(min(t_us_comments.min(), 0), t_us_comments.max(), step=100)  # 设置刻度间隔为10
plt.xticks(x_ticks)

# 添加网格
plt.grid(True)

# 显示图形
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

us_file_path = "./youtube_video_data/US_video_data_numbers.csv"
uk_file_path = "./youtube_video_data/GB_video_data_numbers.csv"

# t1 = np.loadtxt(us_file_path,delimiter=",",dtype="int",unpack=True)
#unpack实现转置的效果
t_uk = np.loadtxt(uk_file_path,delimiter=",",dtype="int")

#取出来英国的youtube评论数和喜欢数,第1列是like数，第3列是comment数

t_uk=t_uk[t_uk[:,1]<=250000] #筛选like数小于等于250000的视频
print(t_uk.shape)
t_uk_like=t_uk[:,1] #取出like数列，第一列，作为横轴坐标

t_uk_comment = t_uk[:,-1] #取出评论数列，最后一列，作为纵轴坐标

plt.figure(figsize=(20,8),dpi=80)
#打印横轴纵轴标签
plt.xlabel("like")
plt.ylabel("comment")
plt.scatter(t_uk_like,t_uk_comment)  #画散点图,观察横轴和纵轴的关系

plt.show()
```
