# 1.1折线图绘制

```python
from matplotlib import pyplot as plt
import random

#左闭右开区间，x轴数据
x = range(8)
#y轴数据

y = [17, 17, 18, 15, 11, 11, 13,15]

#传入参数，绘制折线图
plt.plot(x, y)

#绘制图像
plt.show()
```

# 1.2折线的颜色和形状

```python
#color是颜色，linestyle是线型，marker是标记
plt.plot(x, y, color='r', linestyle='--', marker='o')
plt.show()
```

# 1.4设置图片大小和保存图片

```python
#设置x轴和y轴数据
x = range(2,26,2)
y = [random.randint(15,30)for i in x]

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#绘制折线图
plt.plot(x, y, color='r', linestyle='--', marker='o')

#保存绘制后的图片
plt.savefig('C:\\Users\\12695\\PycharmProjects\\test\\day11\\1.4折线图.png')
plt.savefig('C:\\Users\\12695\\PycharmProjects\\test\\day11\\1.4折线图.svg')

#显示图片
plt.show()
```
