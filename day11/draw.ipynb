# 1.1折线图绘制

# 1.5绘制x轴和y轴的刻度

```python
#设置x轴和y轴数据
x = range(2,26,2)
y = [random.randint(15,30)for i in x]

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置x轴刻度
#将 X 轴刻度设置为 x 列表中的值
#plt.xticks(x)
#将 X 轴刻度设置为从 0 到 25 的整数。
#plt.xticks(range(0,26))

#设置y轴刻度
plt.yticks(y)

#构造x轴刻度标签，刻度标签的数量和 x 轴刻度的数量相同
#生成刻度标签：x_tick_label = ["{}：00".format(i) for i int x] 根据 x 的值生成一个包含 "1：00"，"2：00"，... 的列表作为刻度标签。
x_tick_laberl = ["{}:00".format(i) for i in x]
#设置刻度标签：plt.xticks(x, x_tick_label, rotation=45)：将 x 的值用作 X 轴的刻度位置，x_tick_label 作为对应的标签，并且使用 rotation=45 将标签旋转 45 度以便更好显示
plt.xticks(x,x_tick_laberl,rotation=45)

#构造y轴刻度标签
y_tick_label = ["{}°C".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_tick_label)

#绘制折线图
plt.plot(x, y, color='r', linestyle='--', marker='o')

#显示图片
plt.show()
```

# 1.6设置显示中文

```python
#设置x轴和y轴数据
x = range(0,120)
y = [random.randint(10,30)for i in x]

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置中文字体
from matplotlib import font_manager
#'C:\\Windows\\Fonts\\simsun.ttc'是中文字体的路径，size是字体大小
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc',size=18)

#设置x轴标签
plt.xlabel('时间',rotation=45,fontproperties=my_font,color='red')

#设置y轴标签
plt.ylabel('温度' ,rotation = 0,fontproperties=my_font,color='blue')

#设置标题
plt.title('温度变化图',fontproperties = my_font,color='g')

#绘制折线图
plt.plot(x, y, color='y', linestyle='--', marker='o',label='温度')

#显示图片
plt.show()
```

# 1.7一图多线

```python
#设置x轴和y轴数据
y1 = [random.randint(10,30)for i in range(30)]
y2 = [random.randint(10,30)for i in range(30)]

x =range(30)

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#直接在plot()函数中传入多个y轴数据，即可绘制多条折线
plt.plot(x, y1, color='r', linestyle='--', marker='o',label='会员人数')
plt.plot(x, y2, color='b', linestyle='--', marker='o',label='  订单量')

#设置x轴刻度
x_tick_label = ["{}岁".format(i) for i in x]
plt.xticks(x,x_tick_label,fontproperties=my_font,rotation=45)

#设置y轴刻度
y_tick_label = ["{}个".format(i) for i in range(min(min(y1),min(y2)),max(max(y1),max(y2))+1)]
plt.yticks(range(min(min(y1),min(y2)),max(max(y1),max(y2))+1),y_tick_label,fontproperties=my_font)

#设置图例
plt.legend(prop = my_font,loc='upper right')

#设置网格线
plt.grid(alpha=0.5)

#显示图片
plt.show()
```

# 1.8一图多个子图

```python
import numpy as np

#设置x轴和y轴数据
x = np.arange(1, 100)
fig, axes = plt.subplots(2, 2, figsize=(20, 20))
ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]

# 调整子图之间的距离
plt.subplots_adjust(hspace=1, wspace=1)



#绘制子图1
ax1.plot(x, np.sin(x))
ax1.set_title('sin函数', fontproperties=my_font)
ax1.set_xlabel('x', fontproperties=my_font)
ax1.set_ylabel('sin(x)', fontproperties=my_font)

#绘制子图2
ax2.plot(x, np.cos(x))
ax2.set_title('cos函数', fontproperties=my_font)
ax2.set_xlabel('x', fontproperties=my_font)
ax2.set_ylabel('cos(x)', fontproperties=my_font)


#绘制子图3
ax3.plot(x, np.power(x, 2))
ax3.set_title('x^2函数', fontproperties=my_font,size=25)
ax3.set_xlabel('横坐标', fontproperties=my_font,size=50)
ax3.set_ylabel('x^2', fontproperties=my_font)

#绘制子图4
ax4.plot(x, np.log(x))
ax4.set_title('log函数', fontproperties=my_font)
ax4.set_xlabel('x', fontproperties=my_font)
ax4.set_ylabel('log(x)', fontproperties=my_font)

plt.show()


```

# 1.10条形图

```python
#设置x轴和y轴数据
x = ['A', 'B', 'C', 'D', 'E']
y = [random.randint(10,30)for i in range(5)]

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#绘制条形图
plt.bar(x, y, color='r', label='条形图', width=0.5)

#设置x轴刻度
plt.xticks(x,fontproperties=my_font)

#设置图例
plt.legend(prop = my_font,loc='upper right')

#设置网格线
plt.grid(alpha=0.5)

#在条形图上添加数值
for i in range(len(x)):
    plt.text(i, y[i]+0.5, y[i], ha='center', va='bottom', fontproperties=my_font)

#显示图片
plt.show()
```

# 1.12饼图

```python
#各部分标签
labels = ['A', 'B', 'C', 'D', 'E']

#各部分大小
sizes = [random.randint(10,30)for i in range(5)]

#各部分颜色
colors = ['r', 'g', 'b', 'y', 'c']

#各部分explode
explode = [0.1, 0, 0, 0, 0]

matplotlib.rcParams['font.sans-serif'] = ['SimHei']

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#绘制饼图
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)

#设置图例
plt.legend(prop = my_font,loc='upper right')

plt.axis('equal')

#显示图片
plt.show()
```
